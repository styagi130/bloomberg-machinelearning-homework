# Feature Normalization and visualization

```python
def feature_normalization(train, test):
    """Rescale the data so that each feature in the training set is in
    the interval [0,1], and apply the same transformations to the test
    set, using the statistics computed on the training set.

    Args:
        train - training set, a 2D numpy array of size (num_instances, num_features)
        test  - test set, a 2D numpy array of size (num_instances, num_features)
    Returns:
        train_normalized - training set after normalization
        test_normalized  - test set after normalization

    """
    # TODO
    featureMeans = np.mean(train,axis=0)
    featureMax   = np.max(train,axis=0)
    featureMin   = np.min(train,axis=0)
    
    # Fix for constant features
    notConstantFeatures = ~np.all(trainFeatures==trainFeatures[0,:],axis=0)    
    
    # Fix minimum array to accomodate this normal function
    featureMin          = featureMin*notConstantFeatures
    
    normalizedTrainFeatures = (train - featureMin)/(featureMax-featureMin)
    normalizedTestFeatures  = (test  - featureMin)/(featureMax-featureMin)
    
    return normalizedTrainFeatures,normalizedTestFeatures
```

```python
normalizedTrainFeatures,normalizedTestFeatures = feature_normalization(trainFeatures,testFeatures)


outFeatureDims = 1
infeaturesDims = normalizedTrainFeatures.shape[1]
theta          = np.random.normal(0,1,size=(outFeatureDims,infeaturesDims))
```

# Loss/Loss Gradient Calculation

```python
def predict(theta,x):
    return x.dot(theta.T)

def sq_loss(pred, actual):
    return (pred-actual)**2
```

```python
def compute_square_loss(X, y, theta):
    """
    Given a set of X, y, theta, compute the square loss for predicting y with X*theta

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D array of size (num_features)

    Returns:
        loss - the square loss, scalar
    """
    loss = 0 #initialize the square_loss
    #TODO
    y_                 = predict(theta,X)
    numBatchDataPoints = X.shape[0]
    loss  = sq_loss(y_,y).sum()/numBatchDataPoints
    return loss
```

### Regularized L2 loss

```python
def compute_l2_regularized_square_loss(X,y,theta,lambdaReg):
    squareLoss      = compute_square_loss(X,y,theta)
    regularizedTerm = theta.dot(theta.T)[0,0]
    
    return squareLoss+lambdaReg*regularizedTerm
```

### Gradient of square loss

```python
### compute the gradient of square loss function
def compute_square_loss_gradient(X, y, theta):
    """
    Compute gradient of the square loss (as defined in compute_square_loss), at the point theta.

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D numpy array of size (num_features)

    Returns:
        grad - gradient vector, 1D numpy array of size (num_features)
    """
    #TODO
    y_            = predict(theta,X)
    
    chainRuleDiff = (y_-y).dot(X)
    numBatchDataPoints = X.shape[0]
    
    gradient = 2*(chainRuleDiff.sum(axis=0))/numBatchDataPoints
    return gradient
```

## Regularized Square Loss gradient

```python
def compute_regularized_square_loss_gradient(X, y, theta, lambdaReg):
    """
    Compute the gradient of L2-regularized square loss function given X, y and theta

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        theta - the parameter vector, 1D numpy array of size (num_features)
        lambda_reg - the regularization coefficient

    Returns:
        grad - gradient vector, 1D numpy array of size (num_features)
    """
    squarLossGradient      = compute_square_loss_gradient(X, y, theta)
    regularizationGradient = 2*lambdaReg*theta
    return squarLossGradient+regularizationGradient
```

### Batch gradient descent

```python
def batch_grad_descent(X, y,theta, alpha=0.1, numIter=1000, checkGradient=False):
    """
    In this question you will implement batch gradient descent to
    minimize the square loss objective

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        alpha - step size in gradient descent
        num_iter - number of iterations to run
        check_gradient - a boolean value indicating whether checking the gradient when updating

    Returns:
        theta_hist - store the the history of parameter vector in iteration, 2D numpy array of size (num_iter+1, num_features)
                    for instance, theta in iteration 0 should be theta_hist[0], theta in ieration (num_iter) is theta_hist[-1]
        loss_hist - the history of objective function vector, 1D numpy array of size (num_iter+1)
    """
    numInstances, numFeatures = X.shape[0], X.shape[1]
    thetaHist                 = np.zeros((numIter, *theta.shape))  #Initialize theta_hist
    lossHist                  = np.zeros(numIter) #initialize loss_hist
    gradientHist              = np.zeros((numIter, *theta.shape))
    
    for i in range(numIter):
        loss            = compute_square_loss(X,y,theta)
        gradients       = compute_square_loss_gradient(X,y,theta)
        if loss == np.inf or loss == np.nan:
            print (f"Reached loss of {loss} at epoch: {i}")
            break
        theta           = theta - alpha*gradients
        
        thetaHist[i]    = theta
        lossHist[i]     = loss
        gradientHist[i] = gradients
    return thetaHist[:i],lossHist[:i],gradientHist[:i]
```

```python
def plotLossGradientTheta(thetaHist,lossHist,gradientHist):

    fig = plt.figure(figsize=(10,10))

    axisLoss     = fig.add_subplot(311)
    axisTheta    = fig.add_subplot(312)
    axisGradient = fig.add_subplot(313)

    xAxisValLoss = np.arange(lossHist.shape[0])
    xAxisValGrad = np.arange(thetaHist.shape[-1])

    axisLoss.plot(xAxisValLoss,lossHist)
    axisLoss.set(title="Loss Value with epochs in batch gradient descent")


    for epoch in range(0,thetaHist.shape[0],5):
        axisTheta.plot(xAxisValGrad,thetaHist[epoch].T,label=f"step{epoch}")
    axisTheta.set(title="Theta Values with epochs in batch gradient descent")
    #axisTheta.legend(loc='upper right')

    for epoch in range(0,gradientHist.shape[0],5):
        axisGradient.plot(xAxisValGrad,gradientHist[epoch].T,label=f"step{epoch}")

    axisGradient.set(title="Gradient Values with epochs in batch gradient descent")
    #axisGradient.legend(loc='upper right')
    
    fig.suptitle("Change of various model variables over time")
    return fig
```

```python
theta          = np.random.normal(0.5,0.5,size=(outFeatureDims,infeaturesDims))
thetaHist,lossHist,gradientHist = batch_grad_descent(normalizedTrainFeatures,trainLabels,theta,alpha=0.00001)
fig = plotLossGradientTheta(thetaHist,lossHist,gradientHist)
```

```python
def batch_grad_descentL2(X, y,theta, alpha=0.1, numIter=1000, checkGradient=False,l2Regularization=2):
    """
    In this question you will implement batch gradient descent to
    minimize the square loss objective

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        alpha - step size in gradient descent
        num_iter - number of iterations to run
        check_gradient - a boolean value indicating whether checking the gradient when updating

    Returns:
        theta_hist - store the the history of parameter vector in iteration, 2D numpy array of size (num_iter+1, num_features)
                    for instance, theta in iteration 0 should be theta_hist[0], theta in ieration (num_iter) is theta_hist[-1]
        loss_hist - the history of objective function vector, 1D numpy array of size (num_iter+1)
    """
    numInstances, numFeatures = X.shape[0], X.shape[1]
    thetaHist                 = np.zeros((numIter, *theta.shape))  #Initialize theta_hist
    lossHist                  = np.zeros(numIter) #initialize loss_hist
    gradientHist              = np.zeros((numIter, *theta.shape))
    
    for i in range(numIter):
        loss            = compute_l2_regularized_square_loss(X,y,theta,l2Regularization)
        gradients       = compute_regularized_square_loss_gradient(X,y,theta,l2Regularization)
        if loss == np.inf or loss == np.nan:
            print (f"Reached loss of {loss} at epoch: {i}")
            break
        theta           = theta - alpha*gradients
        
        thetaHist[i]    = theta
        lossHist[i]     = loss
        gradientHist[i] = gradients
    return thetaHist[:i],lossHist[:i],gradientHist[:i]
```

```python
theta          = np.random.normal(0.5,0.5,size=(outFeatureDims,infeaturesDims))
thetaHist,lossHist,gradientHist = batch_grad_descentL2(normalizedTrainFeatures,trainLabels,theta,alpha=0.00001,l2Regularization=5)
fig = plotLossGradientTheta(thetaHist,lossHist,gradientHist)
```

```python
### Stochastic Gradient Descent
def stochastic_grad_descent(X, y,theta, alpha=0.1, lambdaReg=1, numIter=1000):
    """
    In this question you will implement stochastic gradient descent with a regularization term

    Args:
        X - the feature vector, 2D numpy array of size (num_instances, num_features)
        y - the label vector, 1D numpy array of size (num_instances)
        alpha - string or float. step size in gradient descent
                NOTE: In SGD, it's not always a good idea to use a fixed step size. Usually it's set to 1/sqrt(t) or 1/t
                if alpha is a float, then the step size in every iteration is alpha.
                if alpha == "1/sqrt(t)", alpha = 1/sqrt(t)
                if alpha == "1/t", alpha = 1/t
        lambda_reg - the regularization coefficient
        num_iter - number of epochs (i.e number of times) to go through the whole training set

    Returns:
        theta_hist - the history of parameter vector, 3D numpy array of size (num_iter, num_instances, num_features)
        loss hist - the history of regularized loss function vector, 2D numpy array of size(num_iter, num_instances)
    """
    numInstances, numFeatures = X.shape[0], X.shape[1]
    thetaHist                 = np.zeros((numIter,numInstances, *theta.shape))  #Initialize theta_hist
    lossHist                  = np.zeros((numIter,numInstances)) #initialize loss_hist
    gradientHist              = np.zeros((numIter,numInstances, *theta.shape))
    #TODO
    for epoch in range(numIter):
        zippedData = [d for d in zip(X,y)]
        np.random.shuffle(zippedData)
        
        epochThetaHist                 = np.zeros((numInstances,*theta.shape))  #Initialize theta_hist
        epochLossHist                  = np.zeros(numInstances) #initialize loss_hist
        epochGradientHist              = np.zeros((numInstances, *theta.shape))

        for i,(feature,label) in enumerate(zippedData):
            feature         = np.expand_dims(feature, axis=0)
            loss            = compute_l2_regularized_square_loss(feature,label,theta,lambdaReg)
            gradients       = compute_regularized_square_loss_gradient(feature,label,theta,lambdaReg)

            if loss == np.inf or loss == np.nan:
                print (f"Reached loss of {loss} at epoch: {i}")
                break
            theta             = theta - alpha*gradients
            
            epochThetaHist[i]    = theta
            epochLossHist[i]     = loss
            epochGradientHist[i] = gradients
        thetaHist[i]    = epochThetaHist
        lossHist[i]     = epochLossHist
        gradientHist[i] = epochGradientHist
    return thetaHist,lossHist,gradientHist
```

```python
theta          = np.random.normal(0.5,0.5,size=(outFeatureDims,infeaturesDims))
thetaHist,lossHist,gradientHist = stochastic_grad_descent(normalizedTrainFeatures,trainLabels,
                                                          theta,alpha=0.00001,
                                                          lambdaReg=2)
```
